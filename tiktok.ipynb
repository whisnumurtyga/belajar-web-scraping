```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
driver = webdriver.Edge()

url="https://www.tiktok.com"
driver.get(url)
driver.execute_script("document.body.style.zoom = '25%';")
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")
print(soup.prettify()[:700])
```

```python
print(soup.find('a',{"tabindex":"-1"}))
```

```python
video_links = soup.find_all('a',{"tabindex":"-1"})
print("number of videos found:", len(video_links), "\n")

# optional: menampilkan hasil scrape menggunakan print
index = 1
for vl in video_links:
  print(f'video #{index}')
  print(vl)
  video_title_thumb = vl.find('img')
  print(video_title_thumb, "\n")
  index += 1
```

```python
# simpan hasil scrape ke csv
df_videos = pd.DataFrame(columns=["Link", "Title", "Thumbnail"])

for vl in video_links:
  link = vl.get('href')

  video_title_thumb = vl.find('img')
  title = video_title_thumb.get('alt')
  thumb = video_title_thumb.get('src')

  new_row = {"Link":link, "Title":title, "Thumbnail":thumb}

  df_videos.loc[len(df_videos)] = new_row

display(df_videos)
```
