```python
#import terlebih dahulu tools yang diperlukan dalam penilitian
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
driver = webdriver.Edge()

url="https://www.tokopedia.com/"
driver.get(url)
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")
print(soup.prettify()[:700])
```

```python
nama_produk = []
harga_produk = []
penjual = []
kota_toko = []
banyak_terjual = []
rating_produk = []

driver = webdriver.Edge()

for i in range(1,11):
    url="https://www.tokopedia.com/search?navsource=&page={}&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=".format(i)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    for a in soup.find_all('div',{"class":"css-llwpbs"}):
        try:
            nama_produk.append(a.find( 'div', {"class":"prd_link-product-name"} ).get_text())
        except AttributeError:
            nama_produk.append(None)
        try:
            harga_produk.append(a.find( 'div', {"class":"prd_link-product-price"} ).get_text())
        except AttributeError:
            harga_produk.append(None)
        try:
            penjual.append(a.find( 'span', {"class":"prd_link-shop-name"} ).get_text())
        except AttributeError:
            penjual.append(None)
        try:
            kota_toko.append(a.find( 'span', {"class":"prd_link-shop-loc"} ).get_text())
        except AttributeError:
            kota_toko.append(None)
        try:
            banyak_terjual.append(a.find( 'span', {"class":"prd_label-integrity"}).get_text())
        except AttributeError:
            banyak_terjual.append(None)
        try:
            rating_produk.append(a.find( 'span', {"class":"prd_rating-average-text"}).get_text())
        except AttributeError:
            rating_produk.append(None)

    time.sleep(3)

tokopedia = pd.DataFrame()
tokopedia['nama_produk'] = nama_produk
tokopedia['harga_produk'] = harga_produk
tokopedia['penjual'] = penjual
tokopedia['kota_toko'] = kota_toko
tokopedia['banyak_terjual'] = banyak_terjual
tokopedia['rating_produk'] = rating_produk

tokopedia.head()
```
